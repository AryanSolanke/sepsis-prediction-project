```python
"""
This module contains a detailed data analysis on the dataset taken from,
https://www.kaggle.com/datasets/salikhussaini49/prediction-of-sepsis/data
Which is a official 'PhysioNet Cardiology Challenge Dataset 2019' for early predicion of Sepsis.

Context
Sepsis is a life-threatening medical emergency. 
It happens when the body's immune system overreacts to an infection and starts attacking its own tissues and organs, 
eventually leading to organ failure and death if not treated immediately.
Because sepsis moves so fast, every hour of delayed treatment increases the risk of death. 
This project uses machine learning to analyze subtle changes in a patient's vital signs and,
lab results to predict this biological crash before it happens.

Content:
Vital signs (columns 1-8)

1. HR: Heart rate (beats per minute)
2. O2Sat: Pulse oximetry (%)
3. Temp: Temperature (Deg C)
4. SBP: Systolic BP (mm Hg)
5. MAP: Mean arterial pressure (mm Hg)
6. DBP: Diastolic BP (mm Hg)
7. Resp: Respiration rate (breaths per minute)
8. EtCO2: End tidal carbon dioxide (mm Hg)

Laboratory values (columns 9-32)

9. BaseExcess: Measure of excess bicarbonate (mmol/L)
10. HCO3: Bicarbonate (mmol/L)
11. FiO2: Fraction of inspired oxygen (FRACTION)
12. pH: N/A
13. PaCO2: Partial pressure of carbon dioxide from arterial blood (mm Hg)
14. SaO2: Oxygen saturation from arterial blood (%)
15. AST: Aspartate transaminase (IU/L)
16. BUN: Blood urea nitrogen (mg/dL)
17. Alkalinephos: Alkaline phosphatase (IU/L)
18. Calcium: (mg/dL)
19. Chloride: (mmol/L)
20. Creatinine: (mg/dL)
21. Bilirubin_direct: Bilirubin direct (mg/dL)
22. Glucose: Serum glucose (mg/dL)
23. Lactate: Lactic acid (mg/dL)
24. Magnesium: (mmol/dL)
25. Phosphate: (mg/dL)
26. Potassium: (mmol/L)
27. Bilirubin_total: Total bilirubin (mg/dL)
28. TroponinI: Troponin I (ng/mL)
29. Hct: Hematocrit (%)
30. Hgb: Hemoglobin (g/dL)
31. PTT: partial thromboplastin time (seconds)
32. WBC: Leukocyte count (count10^3/µL) Fibrinogen (mg/dL) Platelets (count10^3/µL)

Demographics (columns 33-39)

33. Age: Years (100 for patients 90 or above)
34. Gender: Female (0) or Male (1)
35. Unit1: Administrative identifier for ICU unit (MICU)
36. Unit2: Administrative identifier for ICU unit (SICU)
37. HospAdmTime: Hours between hospital admit and ICU admit
38. ICULOS: ICU length-of-stay (hours since ICU admit)
39. Hour: N/A


Outcome (column 40)

40. SepsisLabel For sepsis patients, SepsisLabel is 1 if t≥tsepsis-6 and 0 if t
"""
```

```python
"""
About Feature 'ICULOS': 

ICULOS is the length-of-stay (hours since ICU admit)

It is a time-series feature and therefore values far from mean/median does not mean the values are outliers.
ICULOS must strictly INCREASE overtime for a patient. 
Also value of ICULOS can't be negative.'

Therefore values not increasing with time for that patient, are possibly outliers and will be removed.
"""

df['ICULOS'].describe()
```

```python
df['ICULOS_diff'] = df.groupby('Patient_ID')['ICULOS'].diff()
df[df['ICULOS_diff'] < 0][['Patient_ID','ICULOS']]
```

```python
"""
About Feature 'HospAdmTime': 

HospAdmTime: Hours between hospital admit and ICU admit

It is a time-series feature and therefore values far from mean/median does not mean the values are outliers.
Outliers check condition: Each patient should have only 1 unique HospAdmTime, if not, the values are outliers and will be removed.
"""
df['HospAdmTime'].describe()
```

```python
"""Outliers detection condition implementation."""

df.groupby(by='Patient_ID')['HospAdmTime'].unique()
```

```python
"""Only 1 patient with ID=13777 has HospAdmTime null, as 1/43,336 is negligible, the patient will be removed."""
df[df['HospAdmTime'].isnull()][['HospAdmTime', 'Patient_ID']]
```

```python
"""
2nd Category of features begins here.
This category contains features with 20-40% missing values, and there analysis before pre-processing.
Features: 'Unit-1', 'Unit-2', and 'DBP'.

About Features:
Unit1 and Unit2 are binary and co-related features. If unit1 = 1 then for all such rows unit2 = 0, making one of them redundant.
Decision: Unit2 will be dropped as it is a derived feature(unit2 = unit1 - 1).

Similar checks of 'MAP' and 'SBP' are applied on 'DBP'
"""
```

```python
"""If unit1 = 1 but unit2 !=0 then such values are surely outliers and will be removed."""\

df[(df['Unit1']==1) & (df['Unit2']!=0)][['Patient_ID', 'Unit1', 'Unit2']]
```
